only the simple code of using whisper model and giving it a command



```python
import sounddevice as sd
from scipy.io.wavfile import write
import whisper

# Record audio
fs = 16000
duration = 4

print("üé§ Speak now...")
audio = sd.rec(int(duration * fs), samplerate=fs, channels=1)
sd.wait()
write("command.wav", fs, audio)

# Speech to text
model = whisper.load_model("base")
result = model.transcribe("command.wav")
text = result["text"].lower()

print("üìù You said:", text)

# Command understanding
def understand_command(text):
    if "switch off" in text and "fan" in text:
        return "FAN_OFF"
    elif "switch on" in text and "fan" in text:
        return "FAN_ON"
    else:
        return "UNKNOWN"

# Execute action
def execute_action(action):
    if action == "FAN_ON":
        print("üü¢ Fan is ON")
    elif action == "FAN_OFF":
        print("üî¥ Fan is OFF")
    else:
        print("‚ö†Ô∏è Command not recognized")

action = understand_command(text)
execute_action(action)



#need to download ffmpeg for whisper to work
#also need the model whisper
```
